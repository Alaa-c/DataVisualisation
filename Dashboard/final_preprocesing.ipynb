```python
filling_values = {"Budget (float)":df['Budget (float)'].mean()}
df = df.fillna(value=filling_values)
df
```

```python
df= df.dropna(how="any")
df.isnull().sum()
```

```python
df["profit"]=df["Box office (float)"]-df["Budget (float)"]
df["profit"]
```

```python
df['year']=pd.DatetimeIndex(df['Release date (datetime)']).year
df['year']
```

```python
arr=df['Country'].str.split(', ').to_numpy()
def flatten_list(_2d_list):
    flat_list = []
    # Iterate through the outer list
    for element in _2d_list:
        if type(element) is list:
            # If the element is of type list, iterate through the sublist
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
countries=flatten_list(arr)
countries
```

```python
countries_df=pd.DataFrame(countries,columns=['country'])
countries_df
```

```python
df=df.drop('Directed by', axis=1).join(df['Directed by'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('director'))
```

```python

df.rename(columns={'Directed by':'director'},inplace=True)
df['director']=df['director'].str.strip('"( )\'[]"')
df['director']
```
